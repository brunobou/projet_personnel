```python
Train=pd.read_csv("df_train_dicho_2.csv",index_col=0)
```

```python
Comp = pd.DataFrame(scaler.fit_transform(X),index=X.index,columns=X.columns)
```

```python
ml= ElasticNetCV(l1_ratio=(0.01,0.05,0.1,0.2,0.4,0.5,0.6,0.8,1),n_alphas=1000,cv=10)
```

```python
ml.fit(x_train,y_train)
```

```python
ml.l1_ratio_
```

```python
ml.alpha_
```

```python
ml.score(x_train,y_train)
```

```python
y_train_pred = ml.predict(x_train)
```

```python
y_target_pred= ml.predict(x_target)
```

```python
mean_squared_error(y_train,y_train_pred)
```
